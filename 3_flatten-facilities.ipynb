```python
import json
import pandas as pd
```

```python
with open('facilities.json', 'r') as f:
    stations = json.load(f)
```

```python
# data = [{ 'name': list(station.keys())[0], 'facilities': station[list(station.keys())[0]] } for station in stations]
```

```python
# with open('facilities_flat.json', 'w') as w:
#    json.dump(data, w)
```

```python
categories = sorted(set([facility['name'] for station in stations for facility in station['facilities']]))
```

```python
flattened = []
for station in stations:
    new_station = { 'name': station['station'] }
    for category in categories:
        if category == 'Number of Parking Places for PMR':
            facility_description = next(iter([
                facility['value'] for facility in station['facilities']
                if facility['name'] == category
            ]), '0')
            new_station[category] = int(facility_description.replace('Nombre de places de parking PMR : ', ''))
        else:
            new_station[category] = int(category in [facility['name'] for facility in station['facilities']])
    flattened.append(new_station)
```

```python
df = pd.DataFrame(flattened)
df
```

```python
df.to_csv('facilities_flattened.csv', index=False)
```
